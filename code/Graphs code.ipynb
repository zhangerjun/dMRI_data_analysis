#First Graph

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive/', force_remount=True)
```

```python
df = pd.read_excel ('drive/My Drive/Internship data/Baby_list_all.xlsx')
```

```python
x = df [(df['scan_age'] <41) & (df['scan_age']>=40)]
```

```python
birth_age = x['birth_age']
```

```python
vPreterm = x [(x['birth_age'] < 32)]
vPreterm
```

```python
birth_weight1 = vPreterm['birth_weight']
import numpy as np
```

```python
vPreterm['scan_age'].values
len(vPreterm['scan_age'].values)
```

```python
scan_head_circumference1 = vPreterm['scan_head_circumference']
```

```python
moderateLatePreterm = x[(birth_age >= 32) & (birth_age < 37)]
moderateLatePreterm
```

```python
birth_weight2 = moderateLatePreterm['birth_weight']
```

```python
scan_head_circumference2 = moderateLatePreterm['scan_head_circumference']
```

```python
termBaby = x[birth_age >=37]
termBaby
```

```python
birth_weight3 = termBaby['birth_weight']
```

```python
scan_head_circumference3 = termBaby['scan_head_circumference']
```

```python
#average birth_weight
ave_birth_weight1 = np.mean(birth_weight1)
ave_birth_weight2 = np.mean(birth_weight2)
ave_birth_weight3 = np.mean(birth_weight3)

#average scan_head_circumference
ave_scan1 = np.mean(scan_head_circumference1)
ave_scan2 = np.mean(scan_head_circumference2)
ave_scan3 = np.mean(scan_head_circumference3)
```

```python
import matplotlib.pyplot as plt

plt.figure()
fig, ax1 = plt.subplots(1,1)
heightBar = [len(vPreterm['scan_age'].values),len(moderateLatePreterm['scan_age'].values),len(termBaby['scan_age'].values)]
plt.bar([0,0.25,0.5], heightBar, width=0.2, bottom=None, align='center', color=['red', 'orange', 'blue'], label=['Very Preterm', 'Moderate/Late Preterm', 'Term baby'])
heightBarWeight = [np.mean(birth_weight1), np.mean(birth_weight2), np.mean(birth_weight3)]
plt.bar([1, 1.25, 1.5], heightBarWeight, width=0.2, bottom=None, align='center', color=['red', 'orange', 'blue'])
heightBarHeadCircumference = [np.mean(scan_head_circumference1), np.mean(scan_head_circumference2), np.mean(scan_head_circumference3)]
plt.bar([2, 2.25, 2.5], heightBarHeadCircumference, width=0.2, bottom=None, align='center', color=['red', 'orange', 'blue'])
plt.title('Graph')


x1 = [0.25,1.25, 2.25]
squad = ['Birth age','Birth weight','Scan head circumference']
ax1.set_xticks(x1)
ax1.set_xticklabels(squad, minor=False, rotation=0)

plt.legend()
#error bars
import statistics as stats
ebars = [np.std(vPreterm['birth_weight'].values), np.std(moderateLatePreterm['birth_weight'].values), np.std(termBaby['birth_weight'].values), np.std(vPreterm['scan_head_circumference'].values), np.std(moderateLatePreterm['scan_head_circumference'].values), np.std(termBaby['scan_head_circumference'].values)]
plt.errorbar([1, 1.25, 1.5, 2, 2.25, 2.5], [ave_birth_weight1, ave_birth_weight2, ave_birth_weight3, ave_scan1, ave_scan2, ave_scan3], yerr = ebars, fmt='o', color='black', markersize=2)

plt.savefig('drive/My Drive/Internship data/Full graph', bbox_inches='tight', pad_inches=0)
#plt.bar(0, len(vPreterm['scan_age'].values), width=0.2, bottom=None, align='center')
#plt.xlim(right=3)
#plt.bar(0.25, len(moderateLatePreterm['scan_age'].values), width=0.2, bottom=None, align='center', color='red')
#plt.bar(0.50, len(termBaby['scan_age'].values), width=0.2, bottom=None, align='center')
```
